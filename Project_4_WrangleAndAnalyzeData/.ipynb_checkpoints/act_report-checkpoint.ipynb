&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Owing to the popularity of this trend, we have decided to gather some data and provide few small interesting insights about this trend. To achieve this, we have extracted the tweet data from twitter's database and also used a neural network to identify the dog breed from each tweet!! Here are some of the insights we could gather that you might quite interesting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_we_rate_dogs = pd.read_csv('twitter_archive_master.csv')
```

### Most favourite breed of dogs

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;We have data about every tweet and also have image predictions of each dog. After assessing and cleaning the data, we can now figure out which breed of dog is most loved by people!

```python
df_favorite = df_we_rate_dogs.groupby('p1')['favorite_count'].sum().reset_index()
df_favorite = df_favorite.sort_values('favorite_count',ascending=False)
fav_count = df_favorite.head(10).favorite_count
breed = df_favorite.head(10).p1
fig,ax = plt.subplots()
plot = plt.bar(breed,fav_count)
plt.xticks(rotation='vertical')
plt.xlabel('Dog Breed')
plt.ylabel('Favourite Count')
plt.title('Favorites by Breed');
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The above bar plot has the top 10 favourite breeds. From the above plot, it can be seen that the Golden retriever is the most loved breed of dog! It leads the 2nd most favourite breed, the Labrador by a large number. The labrador is followed closely by Pembroke.

### Most popular names of dogs

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;It would also be interesting to see what is the most popular name in the dog community. Most users tweet the name of the dog as well and we have gathered this information and have some interesting results below

```python
dog_names = df_we_rate_dogs.name.value_counts()
plt.barh(dog_names.keys()[0:14],dog_names.values[0:14]);
plt.xlabel('Count')
plt.ylabel('Dog Names')
plt.title('Most Popular Dog Names');
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;So from the above horizontal bar plot we can see that Lucy and Charlie are the most popular dog names!! Cooper and Oliver are also not left far behind!

### Retweet Count and Favourite Count
We also have information about the count of retweets of each tweet as well as the number of times a tweet was liked. We can now answer the question if the most retweeted tweets are also the most favorite ones?

```python
plt.scatter(df_we_rate_dogs.retweet_count,df_we_rate_dogs.favorite_count)
plt.xlabel('Retweet Count')
plt.ylabel('Favorite Count')
plt.title('Retweet Count vs Favorite Count');
```

```python
a = str(df_we_rate_dogs['retweet_count'].corr(df_we_rate_dogs['favorite_count']))
print("Correlation between retweet count and favourite count is " + a)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;So we can see from the scatter plot above that the retweets and favourites are positively correlated, which means if a tweet is liked more, it is also retweeted more!
