## Investigation Overview
> In this investigation, we looked at the Ford GoBike service's rental data and tried to see what factors contribute to higher rentals. Our goal is to understand the impact of features in the dataset that contribute to the rental services.

## Dataset Overview
> We have taken the trip data of November 2019 and the data collected has the following features:
<br>Trip Duration (in seconds)
<br>Start Time and Date
<br>End Time and Date
<br>Start Station ID
<br>Start Station Name
<br>Start Station Latitude
<br>Start Station Longitude
<br>End Station ID
<br>End Station Name
<br>End Station Latitude
<br>End Station Longitude
<br>Bike ID
<br>User Type (Subscriber or Customer – “Subscriber” = Member or “Customer” = Casual)
<br>Bike Share for All (Bay Area residents who qualify for CalFresh, SFMTA Lifeline Pass, or PG&E CARE utility discount are eligible to join our **Bike Share for All program** for 5 dollars for the first year)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_fordgobike = pd.read_csv('201911-baywheels-tripdata.csv')
df_fordgobike.start_time = pd.to_datetime(df_fordgobike.start_time)
df_fordgobike.end_time = pd.to_datetime(df_fordgobike.end_time)
df_fordgobike.start_station_id = df_fordgobike.start_station_id.astype(str)
df_fordgobike.end_station_id = df_fordgobike.end_station_id.astype(str)
df_fordgobike.start_station_id = df_fordgobike.start_station_id.replace('nan',np.nan)
df_fordgobike.end_station_id = df_fordgobike.end_station_id.replace('nan',np.nan)
df_fordgobike.drop(columns=['rental_access_method'],inplace=True)
df_fordgobike.dropna(inplace=True)
df_fordgobike['start_hour'] = df_fordgobike.start_time.apply(lambda x : x.hour)
df_fordgobike['end_hour'] = df_fordgobike.end_time.apply(lambda x : x.hour)
bins_day_ticks = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
df_fordgobike['start_day'] = df_fordgobike.start_time.apply(lambda x : bins_day_ticks[x.dayofweek])
df_fordgobike['end_day'] = df_fordgobike.end_time.apply(lambda x : bins_day_ticks[x.dayofweek])
df_fordgobike = df_fordgobike[df_fordgobike.start_station_longitude<-80]
def location(row):
    if row['start_station_longitude']<-122.35:
        return 'San_Francisco'
    elif row['start_station_longitude']<-122.2:
        return 'Oakland'
    elif row['start_station_longitude']<-121.8:
        return 'San_Jose'
    else:
        return 'Other'
    
df_fordgobike['location'] = df_fordgobike.apply(lambda row: location(row),axis=1)
```

## Distribution of the Duration of Bike Rental

> From the data, we can see that most rentals lie below 3000 seconds (50 minutes) but there are a few cases where the bikes have been rented for very long durations. A logarithmetic representation of the duration shows this well. Below, we have plots on both axis - Normal scale & Logarthimetic scale.

```python
bin_edges_normal = np.arange(0,3200,100)
fig,ax1 = plt.subplots(figsize = [10,4])
ax1 = plt.hist(data=df_fordgobike,x='duration_sec',range=[0,3000],bins=bin_edges_normal,color='teal')
plt.grid()
plt.xlabel('Bike Rental Duration (in seconds)');
plt.ylabel('Bike Rental Count');
plt.title('Distribution of Duration of Bike rental in normal scale')

fig,ax2 = plt.subplots(figsize = [10,4])
bin_edges_log = 10 ** np.arange(0.8, np.log10(df_fordgobike.duration_sec.max())+0.1, 0.1)
ax2 = plt.hist(df_fordgobike.duration_sec, bins = bin_edges_log,color='teal')
plt.xscale('log')
plt.grid()
tick_locs = [10, 30, 100, 300, 1000, 3000,10000,30000,100000]
plt.xticks(tick_locs, tick_locs)
plt.xlabel('Bike Rental Duration (in seconds)');
plt.ylabel('Bike Rental Count');
plt.title('Distribution of Duration of Bike rental in Logarithmetic scale');
```

## Distribution of Bike Rental with Days and Hours

> The bike rentals are more common during weekdays and in particular, during the rush hours of 7Am to 9AM and 4PM to 6PM. We can also observe a trend where most rentals in the weekends happen mostly in the afternoon around 12PM to 2PM.  

```python
df_fordgobike['start_day'] = pd.Categorical(df_fordgobike['start_day'], ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
data_counts = df_fordgobike.groupby(['start_hour', 'start_day']).size()
data_counts = data_counts.reset_index(name='count')
data_counts = data_counts.pivot(index = 'start_hour', columns = 'start_day', values = 'count')
plt.figure(figsize = [15, 10])
sb.heatmap(data_counts,cmap='Greens',cbar_kws={'label': 'Bike Rental count'});
plt.xlabel('Day of the Week')
plt.ylabel('Hour of the day')
plt.title('Bike Rentals vs Hour & Day');
```

## Distribution of Customer Types and Geographic Locations

> From our data, we can see that 77% of the users are subscribers and casual customers are just 23%. Moreover, we can see that bike rentals are from 3 areas of California - San Francisco, Oakland and San Jose. San Francisco has the highest number of users (around 76%) followed by Oakland (with 19%) and finally San Jose (with 5%). The ratio of subscribers over casual customers in San Francisco, Oakland and San Jose are around 3.2, 3.7 and 6.42. This shows San Jose has very less number of casual customers

```python
fig, ax = plt.subplots(figsize = (15,10))
ax = sb.countplot(data = df_fordgobike,x = "location",  palette = "viridis", hue = "user_type");
for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate('{:.1f}%'.format(100.*y/len(df_fordgobike)), (x.mean(), y), 
            ha='center', va='bottom')
plt.xlabel('Location in California')
plt.ylabel('Bike Rental count')
plt.title('Distribution of Subscribers and Casual Customers in various locations');
plt.legend(title='User Type',labels=['Casual Customers', 'Subscribers']);
```

## Distribution of Rentals with Customer Type and Duration of Rental

> Trends from the data shows that subscribers usually rent the bikes for shorter periods of time. Casual customers on the other hand happen to rent the bike for longer durations.

```python
bin_edges = 10 ** np.arange(0.8, np.log10(df_fordgobike.duration_sec.max())+0.1, 0.1)
fig, ax = plt.subplots(figsize = (15,10))
for i,user in enumerate(df_fordgobike.user_type.unique()):
    df_data = df_fordgobike[df_fordgobike.user_type == user]
    ax =plt.hist(df_data.duration_sec, bins = bin_edges,histtype='step')
    plt.xscale('log')
    tick_locs = [10, 30, 100, 300, 1000, 3000,10000,30000,100000]
    plt.xticks(tick_locs, tick_locs)
plt.ylabel('Bike Rental count')
plt.xlabel('Bike Rental Duration (in seconds)')
plt.grid()
plt.legend(title='Customer Type',labels=['Casual Customer','Subscriber'])
plt.title('Distribution of Bike Rental Duration vs Customer Types');
```

## Distribution of Rentals with Customer Type and Day & Hour of Rental

> Trends from the data shows that subscribers usually rent the bikes during the peak hours and during weekdays. Casual customers on the other hand, though they rent bikes on rush hours in weekdays, they also happen to contribute to bike rentals during odd hours and weekends more than subscribers do.

```python
df_fordgobike['start_day'] = pd.Categorical(df_fordgobike['start_day'], ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
for i,user in enumerate(df_fordgobike.user_type.unique()):
    df_data = df_fordgobike[df_fordgobike.user_type == user]
    data_counts = df_data.groupby(['start_hour', 'start_day']).size()
    data_counts = data_counts.reset_index(name='count')
    data_counts = data_counts.pivot(index = 'start_hour', columns = 'start_day', values = 'count')
    plt.figure(figsize = [12, 4])
    sb.heatmap(data_counts,cmap='Greens',cbar_kws={'label': 'Bike Rental count'});
    plt.xlabel('Day of the Week')
    plt.ylabel('Hour of the day')
    plt.title('Distribution of Rentals by ' + user.replace('Customer','Casual Customers') + ' and Day & Hour of Rental');
```

The interesting aspect is that longer rental durations are contributed by San Francisco.

## Distribution of Rentals with Geographic location and Hour & Day

> From the data, we can see that San Francisco follows the usual trend of distribution with day and hour of the week, with more users during weekdays and during rush hours. This relationship is weaker with Oakland and is the weakest with San Jose which has users relatively spread accross the week and time of the day in comparison to San Francisco and Oakland. 

```python
fig, ax = plt.subplots(figsize = (10,5))
sb.countplot(data = df_fordgobike,x = "location",  palette = "viridis", hue = "start_day",hue_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']);
plt.xlabel('Location in California')
plt.ylabel('Bike Rental count')
plt.title('Distribution of Bike usage with Days and Locations');
plt.legend(title='Day of the Week');
fig, ax = plt.subplots(figsize = (15,4))
bins_hour = np.arange(0.5,24.5,1)
for i,locn in enumerate(df_fordgobike.location.unique()):
    df_data = df_fordgobike[df_fordgobike.location == locn]
    ax = plt.hist(df_data.start_time.apply(lambda x : x.hour),bins=bins_hour,histtype='step')
    plt.xticks(np.arange(0,24,1));
plt.title('Distribution of Bike usage with Hour of the Day and Locations')
plt.legend(title='Location',labels=list(df_fordgobike.location.unique()))
plt.xlabel('Hour of the Day')
plt.ylabel('Bike Rental count');
```

## Overall Summary
> From the analysis of the Ford GoBike data, we can say that most of the rentals are around 500 seconds and very few above 3000 seconds. Moreover, the service can be seen availed from San Francisco, Oakland and San Jose in California. San Francisco has the highest number of users (around 76%) followed by Oakland (with 19%) and finally San Jose (with 5%). Most rentals happen around the peak hours of 7AM to 9AM and 4PM to 6PM and are more during the weekdays than on weekends. But this relationship is stronger for San Francisco and is weaker in Oakland and is weakest in San Jose. It can be seen that most users are Subscribers and a very few are casual customers. The subscribers use the services primarily during peak hours and on weekdays, thus supporting the statement made above. Casual customers on the other hand use the services more during weekends and also during odd hours. Surprisingly though, San Francisco has a higher percentage of casual customers using the rental service than subscribers in comparison to the Oakland and San Jose area.
